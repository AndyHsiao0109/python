# 錯誤處理
try: <br>
  程式碼 <br>
except: <br>
  例外的時候處理動作 <br>
else: <br>
  當沒有進入例外(程式正確)的處理動作 <br>
finally: <br>
  不管正確例外，都會執行的動作 <br>

```python
try:
  a=10
  b=20
  if a<b: #如果a<b
    print(n) #印出n
except: 
  print("錯誤產生")
```

```python
try:
  a=10
  b=20
  if a<b:
    print(n)
except NameError:
  print("NameError錯誤產生")
except:
  print("其他錯誤產生")
```

```python
try:
  a=10
  b=20
  if a<b:
    print(n)
except NameError:
  print("NameError錯誤產生")
except:
  print("其他錯誤產生")
else:
  print("沒有任何錯誤產生")
finally:
  print("不管程式正確錯誤都會執行的部分")
```
