#Chap 流程控制
## if else , elif

```python
a=eval(input("請輸入數字a:")) #eval會把字串，轉換成數字進行計算
b=eval(input("請輸入數字b:")) 
if a>b : #如果a大於b
  print("a>b") 
elif a==b : #或是a等於b
  print("a=b")
else: #其餘
  print("b>a")
```

##成績評鑑

```python
grade=87
print("成績評鑑：")
if grade >= 90: #如果grade大於90
    print("優等")
elif grade >=80: #或是grade大於80
    print("甲等")
elif grade >=70: #或是grade大於70
    print("乙等")
elif grade >=60: #或是grade大於等於60
    print("丙等")
else: #其餘
    print("不及格")
```

#for 迴圈 
##尾端一定要加上冒號

```python
for i in range(10):  #range(10): 0, 1, 2...9
  print(i)
```

```python
for i in range(0,10,1):  #range(開始, 結束(不包含10),每次進1步)
  print(i)
```

求1+2+3+...10的總和

```python
sum=0
for count in range(1,11,1): #1到10,不包含11
  sum=sum+count

print(sum)
```

```python
for c in "I am a boy": #c的內容
  print(c) 
```

```python
for score in [90,100,88,77]: #score的內容
  print(score)
```

```python
for s in ["john","mary","tom","simpson"]: #s的內容
  print(s)
```

while迴圈

```python
i=0
sum=0
while i<11: #如果i小於11
  sum=sum+i #sum等於sum加上i
  i+=1 #i=i+1
print(sum)
```

##亂數

```python
import random #匯入random模組
for i in range(0,10): #i的範圍
    print(random.randint(1,9)) #亂數產生的數值為1到9
```
