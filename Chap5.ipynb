#圓面積計算函式

```python
import math #匯入math套件
def circle_area(radius):
  area = radius * radius * math.pi
  return area

print("半徑10, 圓面積為：",circle_area(10))
print("半徑5, 圓面積為：",circle_area(5))
print("半徑20, 圓面積為：",circle_area(20))
```
