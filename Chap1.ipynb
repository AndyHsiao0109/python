## 註解

```python
a=5 #a是5 #單行註解
b=3 #b是3
print(a+b) #顯示a+b
```

```python
'''
這是多行註解
這是多行註解
'''
a=5 #a是5
b=3 #b是3
print(a+b) #顯示a+b
```

#運算式

```python
#透過第三變數c，交換a,b的值
a=5 #a是5
b=3 #b是3
c=a #c是5
a=b #a是3
b=c #b是5
print(a) #顯示a
print(b) #顯示b
```

```python
#不透過第三變數，交換a,b的值
a=5 #a是5
b=3 #b是3
a=a+b #a是8
b=a-b #b是5
a=a-b #a是3
print(a) #顯示a
print(b) #顯示b
```

```python
#Python版的變數交換
a=5 #a是5
b=3 #b是3
a,b=b,a #a,b的值換成b,a
print(a) #顯示a
print(b) #顯示b
```

#變數
##變數的平行設定

```python
x,y,z=3,4,5 #x,y,z是3,4,5
print(x) #顯示x
print(y) #顯示y
print(z) #顯示z
```

```python
x,y,z=3,4,5
x+=1 #x=x+1
y*=2 #y=y*2
z**=3 #z=z**3
print(x,y,z) #顯示x,y,z
```

#變數的應用

```python
import math #匯入math模組
r=10 #r是10
print(r*r*math.pi) #顯示10*10*圓周率
```

```python
import math #匯入math模組
r=eval(input("請輸入圓的半徑")) #eval:計算輸入的值
print(r * r * math.pi) #顯示10*10*圓周率
```

#資料型別

```python
x=254 #x是254
print(type(x)) #顯示x的資料類別
y=254.0 #x是254.0
print(type(y)) #顯示y的資料類別
z="write" #z是字串wrtie
print(type(z)) #顯示z的資料類別
print(type(x==y)) #顯示x==y的資料類別
```

#算術符號

```python
x=10 #x是10
y=3 #y是3
print(x/y) #求值
print(x//y) #求商
print(x%y) #求餘數
```

## 算術符號的順序

() > ** > * / > + -

```python
x=(5*(1+2)**2-3+4/2) #()括號>*次方>*/乘除>+-加減
print(x) #顯示x
```

#型態轉換

```python
x="must" #x是字串must
print(10*x) #顯示字串must*10次
print("10"+x) #顯示字串10加上must
print(10,x) #顯示10加上must
```

```python
x=12
print(type(x)) #顯示x的資料類別
y=float(x)+0.5 #強制型別轉換
print(type(y)) #顯示y的資料類別
print(y)
print(int(y))
print(int("12"))
print(float("12.5"))
print(eval("5.5*2"))
```

#字串

```python
print("This"+"is"+"a"+"book") #用+沒有空格
print("This","is","a","book") #用,有空格
print("This")
print("is")
print("a")
print("book")
```

```python
print("This {} a book".format("is")) #利用字串的str.format()，在{ }中替代變數的位置
x=123456.7819
print("{:.2f}".format(x)) #小數點.後兩位
print("{:,.2f}".format(x)) #,小數點前用,分開
```

```python
print("This %10s a book%-10s" % ("is","!")) #字串占10格，靠右2格
print("%7.3f" % (687.4333))
```
